```python
#4. Use PyPDF2 to extract the text out of each document (fox)

foxpdf = PdfReader('fox2.pdf')
foxpage = len(foxpdf.pages)
foxtext = ''

for foxpages in range(foxpage):
    text = foxpdf.pages[foxpages].extract_text()
    foxtext += text + ' '
 
print(foxtext)
```

```python
#4.5 (reuters)
reuterspdf = PdfReader('reuters.pdf')
reuterspage = len(reuterspdf.pages)
reuterstext = ''

for page in range(reuterspage):
    text = reuterspdf.pages[page].extract_text() #page is page number
    reuterstext += text + ' '
 
print(reuterstext)
```

```python
#7. Using py-readability-metrics complete 2-3 different readability scores
#a. Discuss what the chosen metric is measuring
#b. Compare the scores from both articles
import readability
from readability import Readability

#Automaded Redability Index (1)
#This metric is measuring The formula calculates the grade level of a text sample based on sentence length and number of unfamiliar words. 
#The Spache Formula considers “unfamiliar words” as words that 3rd grade and below do not recognize.
#ARI does on a factors of characters per word instead of the usual syllables per word

reuter = Readability(reuterstext)
fox = Readability(foxtext)

arireuter = reuter.ari()
print(f'This is the Automated Redability Index score for reuters:{arireuter.score}')
print(f'This is the Automated Redability Index grade level for reuters:{arireuter.grade_levels}')
print(f'This is the Automated Redability Index ages for reuters:{arireuter.ages}')

arifox = fox.ari()
print(f'This is the Automated Redability Index score for fox:{arifox.score}')
print(f'This is the Automated Redability Index grade level for fox:{arifox.grade_levels}')
print(f'This is the Automated Redability Index ages for fox:{arifox.ages}')
```

```python
#Redability Score (2)
#The Spache Formula considers “unfamiliar words” as words that 3rd grade and below do not recognize.

sfox = fox.spache()
sreuter = reuter.spache()

print(f'This is the SMOG score for fox:{sfox.score}')
print(f'This is the SMOG grade level for fox:{sfox.grade_level}')

print(f'This is the SMOG score for fox:{sreuter.score}')
print(f'This is the SMOG grade level for fox:{sreuter.grade_level}')
```
